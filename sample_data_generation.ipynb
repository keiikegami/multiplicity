今想定する状況は、Pop, Distともに効用関数に入り、$\alpha > 0$で$\beta < 0$である。その上で以下のように場合分けする。

- $\delta > 0$、すなわちstrategic complement。$\alpha$が与えられた時、$\beta$がある程度大きくないと複数均衡は生まれない。
- $\delta < 0$、すなわちstrategic substitute。$\alpha$が与えられた時、$\beta$がある程度小さくないと複数均衡は生まれない。

PopとDistは乱数で発生させる。BR functionsのnotebookで確認したように、PopやDistの値、そしてパラメータの設定によっては均衡が一つしかないパターンもあるし、なんなら相手の行動がどのように予測できていようと参入or退出を確率1で選んだほうがいい、みたいな状況すら存在する。

なので、パラメータを設定したら、いろいろなパターンがうまく出るように乱数を発生させないとね。
